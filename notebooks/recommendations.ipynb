```python
import pandas as pd
import numpy as np
```

```python
liked_books = ["5", "81605", "62999", "36776", "60730", "33908"]
```

```python
books_data = pd.read_csv("../data/raw/books_data.csv")
books_rating = pd.read_csv("../data/raw/Books_rating.csv")
```

```python
books_data.dropna(inplace=True)
```

```python
books_rating["review/score"] = books_rating["review/score"].astype(int)
```

```python
len(books_rating)
```

```python
def rename_title(df, title_column_name):
    df["mod_title"] = df[title_column_name].str.lower().replace("[^a-z0-9 ]", "", regex=True).str.replace("\s+", " ", regex=True)
    #.replace("[^a-z0-9]", "", regex=True)
    #.str.replace("\s+", " ", regex=True)
    return df

books_data= rename_title(books_data, "Title")
books_rating = rename_title(books_rating, "Title")
```

```python
books_data_reset = books_data.reset_index()
```

```python
books_data_reset
```

```python
title_to_index = books_data.reset_index().set_index('mod_title')['index'].to_dict()
```

```python
books_rating['book_id'] = books_rating['mod_title'].map(title_to_index)
```

```python
books_rating.isnull().sum()
```

```python
books_rating.dropna(subset=["User_id", "review/score", "book_id"], inplace=True)
```

```python
books_rating["book_id"] = books_rating["book_id"].astype(int)
```

```python
books_rating
```

```python
overlap_users = set() #muss ein Set sein, damit duplikate automatisch entfernt werden
```

```python
def check_for_similar_users(row):
    if str(row["book_id"]) in liked_books and row["review/score"] >= 4:
        overlap_users.add(row["User_id"])
```

```python
books_rating.apply(check_for_similar_users, axis=1)
```

```python
recs_lines = []
```

```python
def check_for_recs(row):
    if row["User_id"] in overlap_users:
        recs_lines.append([row["User_id"], row["book_id"], row["review/score"]])
```

```python
books_rating.apply(check_for_recs, axis=1)
```

```python
recs = pd.DataFrame(recs_lines, columns=["User_id", "book_id", "review/score"])
recs["book_id"] = recs["book_id"].astype(str)
```

```python
recs
```

```python
top_recs = recs["book_id"].value_counts().head(10)
top_recs = top_recs.index.values
```

```python
top_recs
```
